````python
import re 
import unicodedata
def rem_hash_user(text:str):
    if pd.isna(text):
        return ""
    hash_user = re.findall(r"(@\w+|#\w+|@\d+|#\d+)", text)
    hash_user= " ".join(hash_user)
    return hash_user

df_2["hash_user"] = df_2["Text"].apply(rem_hash_user)

df_2.sort_values("Text",ascending=True).head(6)

````

````python
import re
import unicodedata

def clean(text: str):
    if pd.isna(text):
        return ""
    # 1. Normalize Unicode (fix weird chars like Â )
    text = unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf-8")

    # 2. Lowercase
    text = text.lower()

    # 3. Remove mentions (@usernames)
    text = re.sub(r"@\w+", "", text)

    # 4. Remove hashtags (optional: keep if needed)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"#", "", text)
    text = re.sub(r"rt :", "", text)
    text = re.sub(r'\b(?:a|[a-z]*\d+[a-z]*)\b', '', text, flags=re.IGNORECASE)
    
    # 5. Remove line breaks and extra whitespace
    text = re.sub(r"\s+", " ", text).strip()

    # 6. Remove special characters (keep only alphanumeric and space)
    text = text.replace('"', "")
    text=text.strip()
    return text

df_2["Text"] = df_2["Text"].apply(clean)

df_2.sort_values("Text",ascending=True).head(5)

````

- Tokenize the text - word_tokenize
 - Remove stop words - stopwords.words('english')
 - Lemmatize the tokens- WordNetLemmatizer().lemmatize()
 - Join the tokens back into a string

````python
import nltk
nltk.download('punkt_tab')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

def preprocess_text(text:str):
    if type(text)==float:
        return ""
    else:
        tokens=word_tokenize(text=text)
        rem_stopword=[token for token in tokens if token not in stopwords.words('english')]
        lematizer=WordNetLemmatizer()
        lematize=[lematizer.lemmatize(word) for word in rem_stopword]
        sent=" ".join(lematize)
        return sent
new_df["processed_Text"]=new_df["Text"].apply(preprocess_text)
# print(preprocess_text(text="hello brother whatcha you doing , you nigga how's your are brother"))
````
